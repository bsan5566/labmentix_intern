#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
data.isnull().sum()
````

Let's drop the null values

````python
# Drop rows with missing descriptions
data.dropna(subset=['description'], inplace=True)

# Fill missing age_certification values with the mode
age_mode = data['age_certification'].mode()[0]
data['age_certification'].fillna(age_mode, inplace=True)

# Fill missing seasons values with 0
data['seasons'].fillna(0, inplace=True)

# Drop rows with missing IMDb IDs
data.dropna(subset=['imdb_id'], inplace=True)
````

````python
# Fill missing values in 'imdb_score' with its mean (rounded to 1 decimal)
imdb_mean = round(data['imdb_score'].mean(), 1)
data['imdb_score'].fillna(imdb_mean, inplace=True)

# Fill missing values in 'imdb_votes' with 0
data['imdb_votes'].fillna(0, inplace=True)

# Fill missing values in 'tmdb_popularity' with its mean (rounded to 2 decimals)
tmdb_popularity_mean = round(data['tmdb_popularity'].mean(), 2)
data['tmdb_popularity'].fillna(tmdb_popularity_mean, inplace=True)

# Fill missing values in 'tmdb_score' with its mean (rounded to 1 decimal)
tmdb_score_mean = round(data['tmdb_score'].mean(), 1)
data['tmdb_score'].fillna(tmdb_score_mean, inplace=True)

# Fill missing values in 'character' with 'unknown'
data['character'].fillna('unknown', inplace=True)
````

# Chart-1:-Box plot of Release_years



````python
# Box plot of release_year
plt.figure(figsize = (8,6))
sns.boxplot(data = data, y = 'release_year')
median = data['release_year'].median()
plt.text(0, median, f'Median: {median:.0f}', color='blue', fontsize=12, ha='center', va='bottom', fontweight='bold')

plt.title('Box plot of Release_year\n', color = 'red')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The Box Plot is an ideal chart type for visualizing the distribution and identifying outliers in a continuous numerical variable — in this case, `release_year.`

## Reasons for Choosing the Box Plot:
1. Shows Distribution Clearly:

 * The box represents the interquartile range (IQR) (middle 50% of data).

 * The line inside the box shows the median release year.

2. Highlights Outliers: Any dots outside the "whiskers" are outlier years, which can help us spot very old or incorrectly entered data (e.g. 1910s, 1920s).

3. Compact and Informative: One chart provides the spread, central tendency, and anomalies in a compact form.

4. Perfect for Time-Related Numeric Columns: release_year is numerical, continuous, and has a wide range (1910s–2020s), making it ideal for box plot visualization.

##### 2. What is/are the insight(s) found from the chart?

1.Most Content Is Recent: The interquartile range (IQR) spans from approximately 1980 to 2020, indicating that the majority of content on Amazon Prime Video was released during this period.

2. Median Release Year: The median (middle line in the box) lies just after 2000, meaning over half the content was released in the 21st century.

3. Older Outliers Exist: There are outliers before 1940, including some even before 1920. These could be_ Classic films,Rare documentaries Or possibly data entry errors

4. Limited Very Recent Outliers: There’s a tight upper whisker, suggesting relatively fewer extremely new releases beyond the IQR (after 2020).

# Chart-6 :- Line chart of seasons and movies released per year

````python
release_counts = data['release_year'].value_counts().sort_index()
total_seasons = data.groupby('release_year')['seasons'].sum()

# Align common years
common_years = release_counts.index.intersection(total_seasons.index)

# Line chart 1: Movies released per year
plt.figure(figsize=(10, 5))
plt.plot(common_years, release_counts.loc[common_years], marker='o', color='blue')
plt.title('Number of Movies Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line chart 2: Total seasons per year
plt.figure(figsize=(10, 5))
plt.plot(common_years, total_seasons.loc[common_years], marker='s', color='green', linestyle='--')
plt.title('Total Seasons per Release Year')
plt.xlabel('Release Year')
plt.ylabel('Total Seasons')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The line chart with subplots was chosen because:

* It clearly shows trends over time, making it ideal for analyzing how the number of movies and TV show seasons have evolved by release year.

* Using two vertically stacked plots allows us to compare related trends (movies vs. seasons) without overcrowding a single plot.

* It’s easy to spot peaks, drops, or steady growth over years.

* A clean layout with shared x-axis improves readability and ensures consistent year comparison.

# Chart-7 :- Bar chart of imdb scores and seasons

````python
data['imdb_score_rounded'] = data['imdb_score'].round()

# Group by rounded IMDb score and calculate average number of seasons
score_season_avg = data.groupby('imdb_score_rounded')['seasons'].mean()

# Bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(score_season_avg.index, score_season_avg.values, color='coral', edgecolor='black')

# Labels & Titles
plt.xlabel('IMDb Score', fontsize=12, fontfamily='serif')
plt.ylabel('Average Number of Seasons', fontsize=12, fontfamily='serif')
plt.title('Average Seasons by IMDb Score', fontsize=16, fontfamily='serif', color='darkred')
plt.xticks(score_season_avg.index, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Adding data labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.1f}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The bar chart was chosen because:

- It clearly displays the average number of seasons across rounded IMDb scores in a comparative format.

- Bar charts are ideal for showing aggregated statistics like averages across categories.

- The use of rounded IMDb scores helps simplify the x-axis and improves readability, avoiding clutter from float values.

- It provides a quick visual interpretation of how critical reception (ratings) may influence the longevity (seasons) of shows.

##### 2. What is/are the insight(s) found from the chart?

From the chart, we can observe:

- Higher IMDb scores (e.g., 7–9) tend to have a greater average number of seasons, indicating that higher-rated shows are more likely to be renewed.
